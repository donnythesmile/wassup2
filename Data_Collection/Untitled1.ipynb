```python
import requests
from bs4 import BeautifulSoup
import csv

url = 'https://www.melon.com/chart/index.htm'
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

chart_list = soup.select('table > tbody > tr')

with open('melon_top100.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['순위', '제목', '가수', '좋아요수'])

    for song in chart_list:
        rank = song.select_one('span.rank').text
        title = song.select_one('div.ellipsis.rank01 > span > a').text
        artist = song.select_one('div.ellipsis.rank02 > a').text
        likes = song.select_one('span.cnt').text.replace(',', '')
        writer.writerow([rank, title, artist, likes])

print("스크래핑 완료. 'melon_top100.csv' 파일을 확인하세요.")
```

```python
import requests
from bs4 import BeautifulSoup
import csv
import os

url = 'https://www.melon.com/chart/index.htm'
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

chart_list = soup.select('table > tbody > tr')

if not os.path.exists('album_images'):
    os.makedirs('album_images')

with open('melon_top100_with_images.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['순위', '제목', '가수', '좋아요수', '앨범 이미지'])

    for song in chart_list:
        rank = song.select_one('span.rank').text
        title = song.select_one('div.ellipsis.rank01 > span > a').text
        artist = song.select_one('div.ellipsis.rank02 > a').text
        likes = song.select_one('span.cnt').text.replace(',', '')
        
        album_image_url = song.select_one('img')['src']
        album_image_filename = f"album_images/{rank}_{title}.jpg"
        with open(album_image_filename, 'wb') as f:
            f.write(requests.get(album_image_url).content)
        
        writer.writerow([rank, title, artist, likes, album_image_filename])

print("스크래핑 완료. 'melon_top100_with_images.csv' 파일과 'album_images' 폴더를 확인하세요.")
```

```python
import requests
from bs4 import BeautifulSoup
import csv
import os

url = 'https://www.melon.com/chart/index.htm'
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

chart_list = soup.select('table > tbody > tr')

if not os.path.exists('album_images'):
    os.makedirs('album_images')

with open('melon_top100_with_images.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['순위', '제목', '가수', '좋아요수', '앨범 이미지'])

    for song in chart_list:
        rank = song.select_one('span.rank').text
        title = song.select_one('div.ellipsis.rank01 > span > a').text
        artist = song.select_one('div.ellipsis.rank02 > a').text
        likes = song.select_one('span.cnt').text.replace(',', '')
        
        album_image_url = song.select_one('img')['src']
        album_image_filename = f"album_images/{rank}_{title}.jpg"
        album_image_path = os.path.abspath(album_image_filename)
        with open(album_image_filename, 'wb') as f:
            f.write(requests.get(album_image_url).content)
        
        writer.writerow([rank, title, artist, likes, album_image_path])

print("스크래핑 완료. 'melon_top100_with_images.csv' 파일과 'album_images' 폴더를 확인하세요.")
```

```python
import requests
from bs4 import BeautifulSoup
import csv

# 웹 페이지에서 데이터 가져오기
url = "https://lol.ps/statistics"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# 필요한 데이터 추출
data = []
table = soup.find('table')
rows = table.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    cols = [col.text.strip() for col in cols]
    data.append(cols)

# CSV 파일로 저장
with open('lol_statistics.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data)
```

```python
import requests
from bs4 import BeautifulSoup
import csv

url = 'https://finance.naver.com/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

top_stocks = soup.select_one('.section_top_lft')

data = []

for item in top_stocks.select_one('.lst_pop > li'):
    rank = item.select_one('.num').get_text(strip=True)
    stock_name = item.select_one('.txt').get_text(strip=True)
    current_price = item.select_one('.num2').get_text(strip=True)
    change = item.select_one('.num3').get_text(strip=True)
    data.append([rank, stock_name, current_price, change])

with open('top_stocks.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['순위', '종목명', '현재가', '전일대비'])
    writer.writerows(data)

print('TOP종목 데이터가 top_stocks.csv 파일로 저장되었습니다.')
```

```python
import requests
from bs4 import BeautifulSoup
import csv

# HTML 코드를 문자열로 저장합니다.
html = '''
<div class="section_sise_top">
    https://finance.naver.com/
</div>
'''

# BeautifulSoup 객체를 생성합니다.
soup = BeautifulSoup(html, 'html.parser')

# 각 섹션의 데이터를 추출합니다.
sections = ['거래상위', '상승', '하락', '시가총액 상위']
data = []

for section_id in ['_topItems1', '_topItems2', '_topItems3', '_topItems4']:
    section_name = sections.pop(0)
    rows = soup.select(f'#({section_id} > tr)')
    for row in rows:
        cols = row.select('th > a, td')
        cols = [col.text.strip() for col in cols]
        data.append([section_name] + cols)

# CSV 파일로 저장합니다.
with open('top_items.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerows(data)
```
