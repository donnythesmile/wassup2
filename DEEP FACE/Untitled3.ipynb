```python
from deepface import DeepFace

# 비교할 두 이미지의 경로
img1_path = 'C:/Users/jsd76/workspace/python/wassup2/DEEP FACE/img1.jpg'
img2_path = 'C:/Users/jsd76/workspace/python/wassup2/DEEP FACE/img2.jpg'

# 두 이미지 간의 유사도 측정
result = DeepFace.verify(
    img1_path,
    img2_path,
    model_name='Facenet',
    detector_backend='mtcnn',
    distance_metric='cosine',  # 거리 측정 방식 설정 (cosine, euclidean, euclidean_l2)
    enforce_detection=False,   # 얼굴 검출 강제 여부 설정
    align=True                 # 얼굴 정렬 수행 여부 설정
)

# 유사도 계산
distance = result['distance']
similarity = (1 - distance) * 100

# 유사도에 따른 결과 출력
if similarity < 80:
    print(f"안면 유사도 검사 결과: {similarity:.2f}%로 해당 사진은 동일인이 아닌 것으로 추정됩니다.")
elif 80 <= similarity < 90:
    print(f"안면 유사도 검사 결과: {similarity:.2f}%로 해당 사진은 동일인이지만 과도한 보정이 적용된 것으로 추정됩니다.")
elif 90 <= similarity < 95:
    print(f"안면 유사도 검사 결과: {similarity:.2f}%로 해당 사진은 동일인이지만 약간의 보정이 적용된 것으로 추정됩니다.")
else:
    print(f"안면 유사도 검사 결과: {similarity:.2f}%로 해당 사진은 동일인임을 인증합니다. 증명사진으로 사용가능합니다.")
```
